```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IRIS1.csv')
```

### Univariant Analysis on Continues Data
- I will take one varable analysis data on one varable
- We do Univariant analysis on continues data

```python
plt.figure(figsize=(6,6))
sns.stripplot(x='sepal_length',data = df, jitter = True) ## By default jitter is True
## jitter controll the variation
## y axis dose not have any meaning
plt.title("Strip Plot Of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

```python
## Swarm Plot
plt.figure(figsize= (6,6))
sns.swarmplot(x = 'sepal_length', data = df)
plt.title("Swarm Plot of Sepal Length")
## in Y axis we do not have any meaning
plt.xlabel("Sepal Length")
plt.show()
```

```python
## Histogram Plot
plt.figure(figsize=(6,6))
sns.histplot(x = 'sepal_length',data = df,kde = True)
plt.title("Hist Plot On Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Count")
plt.show()
```

```python
## Density Plot
plt.figure(figsize = (6,6))
sns.kdeplot(x = 'sepal_length', data = df, bw_adjust=0.5)
plt.title("Density Plot on Sepal Length")
plt.xlabel("Seapal Length")
plt.ylabel("DEnsity")
plt.show()
```

```python
## Rug Plot
plt.figure(figsize = (6,6))
sns.rugplot(x = 'sepal_length', data = df)
plt.title("Rug Plot on Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
```

```python
plt.figure(figsize= (6,6))
sns.lineplot(x = 'sepal_length', y = df.index , data = df)
plt.title("Line Plot for Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Index")
plt.show()
```

```python
plt.figure(figsize= (6,6))
sns.scatterplot(x = 'sepal_length', y = df.index , data = df)
plt.title("Scatter Plot for Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Index")
plt.show()
```
